# ADFS Gen2 Connection Example

1. Create Azure ADLS Gen Account
2. Create Key Vault to store the secrets
3. Create web application through app Registration
4. Store secret Key  in key Valut (Which is generated during the App Registration)
5. IMP : Give permission to register app to access ADLS 
6. Create secret scope in Databricks for azure key valut 
7. create Mount point fo ADLS in Azure data bricks
8. Load data from ADLS and create dataframe
